# Create Series from dictionaries

```python
data = {'a':0,'b':1,'c':3}
s=pd.Series(data)
print(s)


```

#   Accessing Data from Series

```python
s = pd.Series([1,2,3,4], index = ['a','b','c','d'])
print(s[0])    # print using postition
print(s[:3])
print(s[2:4])
print(s['c'])  # print using label
```
